To clean, transform, and prepare the Titanic dataset for training a machine learning model. This includes handling missing values, encoding categorical variables, and selecting relevant features.

ðŸ”¹ Cell 1: Load Required Libraries

```python
import pandas as pd

# Download the Titanic dataset from GitHub and save a local copy
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.to_csv("titanic.csv", index=False)  # Save for future use

# Show first 5 rows
df.head()
```

ðŸ’¬ What this Cell Does:
Imports the pandas library for data analysis.
Loads the Titanic dataset from an online GitHub URL.
Saves a local copy as titanic.csv for reuse later.
Displays the first 5 rows of the dataset to preview the columns.

ðŸ”¹ Cell 2: Dataset Info & Missing Values
ðŸ§  What it does:
Displays column names, data types, and non-null counts.
Shows how many missing values exist in each column.

```python
# Show structure and missing data
df.info()
df.isnull().sum()
```

ðŸ”¹ Cell 3: Handle Missing Values (Age & Embarked)
ðŸ§  What it does:
Fills missing age values with the median age.
Fills missing embarkation ports with the most common one.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

ðŸ”¹ Cell 4: Drop Useless Columns
ðŸ§  What it does:
Removes irrelevant or high-cardinality columns that donâ€™t help in prediction.

```python
# Drop columns we wonâ€™t use in modeling
df.drop(["Cabin", "Ticket", "Name", "PassengerId"], axis=1, inplace=True)
df.head()
```

ðŸ”¹ Cell 5: Encode Categorical Variables
ðŸ§  What it does:
Converts Sex and Embarked into numeric values using one-hot encoding.

```python
# Convert categorical columns into dummy/one-hot variables
df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True)
df.head()
```
